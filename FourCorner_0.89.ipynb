```python
import cv2
import matplotlib.pyplot as plt
import imutils
from imutils.perspective import four_point_transform
import numpy as np
import time
```

```python
def FourCorner(number):
    image = cv2.imread('E:/jpg/'+number+'.JPG')
    docCnt = None
    k=0
    image2=image.copy()
    
    for j in range(0,8):
        if docCnt==None:
            gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
                #blurred = cv2.medianBlur(gray, 43, 0)
            for z in (43,33):
                blurred = cv2.medianBlur(gray, z)
                blurred=cv2.adaptiveThreshold(blurred,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,51,2)
                kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3, 3))
                #blurred = cv2.erode(blurred,kernel)
                blurred = cv2.dilate(blurred,kernel)
                edged = cv2.Canny(blurred, 10, 100)
                cnts = cv2.findContours(edged, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
                cnts = cnts[0] if imutils.is_cv2() else cnts[1]
                print ("number of contours detected ", len(cnts))
                if len(cnts) > 0 :
                    cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
                    c=cnts[0]
                    peri = cv2.arcLength(c, True)
                    area = cv2.contourArea(c)
                    print(area)
                    if  area > 2000000:
                        for a in range(1,10):
                            approx = cv2.approxPolyDP(c, 0.1*a*peri, True)
                        # 如果近似轮廓有四个顶点，那么就认为找到了答题卡
                            if len(approx) ==4:
                                docCnt = approx
                                break
                if docCnt is not None:
                    break
        if docCnt is None and j == 7: 
            k=10
            break
        if docCnt is not None: 
            k=j
            break
        image2=cv2.pyrMeanShiftFiltering(src=image,sp=5,sr=(j+1)*10)
        
    newimage=image.copy()
    #docCnt 为一个np矩阵或者NoneType
    if k!=10:
        for i in docCnt:
    #circle函数为在图像上作图，新建了一个图像用来演示四角选取
            cv2.circle(newimage, (i[0][0],i[0][1]), 50, (255, 0, 0), -1)
        paper = four_point_transform(image, docCnt.reshape(4, 2))
        warped = four_point_transform(gray, docCnt.reshape(4, 2))
    else:paper=image
    return paper,newimage
```

```python
starttime = time.time()
for i in range(1,185):
    print(str(i))
    paper,newimage=FourCorner(str(i))
    cv2.imwrite('E:/o_3c_4/'+str(i)+'_0.jpg', paper)
    cv2.imwrite('E:/o_3c_4/'+str(i)+'_1.jpg', newimage)
endtime = time.time()
duringtime=endtime-starttime
print(duringtime)
```
